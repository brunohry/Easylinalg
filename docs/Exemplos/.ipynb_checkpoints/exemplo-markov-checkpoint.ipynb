# Setas

```julia
s1 = Arrow2D(1,0)
Draw(s1)
```

```julia
s2 = Arrow2D(0,2)
Draw(s2)
```

```julia
Draw((s1+ s2) / 2)
```

```julia
s3 = Arrow3D(1,0,1)
Draw(s3)
```

```julia
s4 = Arrow3D(0,1,-1)
Draw(s4)
```

# Signal

```julia
sg1 = SignalBuildU(x -> cos(100*x), -10:0.1:10)
Draw(sg1)
```

```julia
sg2 = SignalBuildU(x -> 10*sin(x), -10:0.1:10)
Draw(sg2)
```

```julia
Draw((sg1 + sg2) /2)
```

```julia
Draw((sg1 + sg2) - sg1)
```
