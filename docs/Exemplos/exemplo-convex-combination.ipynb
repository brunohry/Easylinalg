```julia
using EasyLinalg
```

# Points

```julia
p1 = Point2D(1, 1)
p2 = Point2D(10, 10)
Draw([p1, p2])
```

```julia
Draw(convex_combination([p1, p2], 0.1))
```

```julia
p_3d_1 = Point3D(1, 1, 1)
p_3d_2 = Point3D(10, 10, 10)
Draw([p_3d_1, p_3d_2])
```

```julia
Draw(convex_combination([p_3d_1, p_3d_2], 0.1),)
```

# Setas

```julia
s1 = Arrow2D(1,0)
s2 = Arrow2D(0,2)
Draw([s1, s2])
```

```julia
Draw(convex_combination([s1, s2], 0.1))
```

```julia
s_3d_1 = Arrow3D(1,0,1)
s_3d_2 = Arrow3D(0,1,-1)
Draw([s_3d_1, s_3d_2])
```

```julia
Draw(convex_combination([s_3d_1, s_3d_2], 0.1))
```

# Signal

```julia
sg1 = SignalBuildU(cos, 0:0.1:10)
sg2 = SignalBuildU(sin, 0:0.1:10)
Draw([sg1, sg2])
```

```julia
Draw(convex_combination([sg1, sg2], 0.33333), true, true)
```
